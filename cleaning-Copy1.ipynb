import os
import re

def rename_images(folder_path):
    # Dictionary to keep track of the original and new names
    name_map = {}
    
    # Regular expression to match the leading number and period
    pattern = re.compile(r'^\d+\.')

    for filename in os.listdir(folder_path):
        if filename.endswith('.jpg'):  # Process only .jpg files
            # Remove the leading number and period
            new_name = pattern.sub('', filename)

            # Check for duplicates and rename if necessary
            original_new_name = new_name
            count = 1
            while new_name in name_map.values():
                new_name = f"{os.path.splitext(original_new_name)[0]}{count}.jpg"
                count += 1

            # Add the original and new name to the dictionary
            name_map[filename] = new_name

    # Rename the files
    for original, new in name_map.items():
        os.rename(os.path.join(folder_path, original), os.path.join(folder_path, new))

    print("Renaming completed.")

     # Specify the folder path
folder_path = 'resturant/data/data'
rename_images(folder_path)
