```python
import os
import pandas as pd
import requests

# Create a directory to store the images
image_folder = 'images_url'
os.makedirs(image_folder, exist_ok=True)

# Specify the path to your CSV file
csv_file_path = 'resturant/cuisine_updated.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Iterate through the 'image_url' column and download images
for index, row in df.iterrows():
    image_url = row['image_url']
    image_name = os.path.basename(image_url)  # Extract the image name from the URL
    image_path = os.path.join(image_folder, image_name)

    try:
        response = requests.get(image_url)
        if response.status_code == 200:
            with open(image_path, 'wb') as file:
                file.write(response.content)
            print(f"Downloaded {image_name}")
        else:
            print(f"Failed to download {image_name}")
    except Exception as e:
        print(f"Error downloading {image_name}: {str(e)}")

print("Image download completed.")
```
